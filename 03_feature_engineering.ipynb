```python
df = pd.read_csv("../data/cleaned/cleaned_procurement.csv")
df.head()
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])

df['Month'] = df['Order_Date'].dt.to_period('M')
```

```python
category_monthly = df.groupby(['Item_Category', 'Month'])['Quantity'].sum().reset_index()

category_monthly['Month'] = category_monthly['Month'].astype(str)

category_monthly.head()
```

```python
category_monthly = category_monthly.sort_values(['Item_Category', 'Month'])
```

```python
# approximate spend using average unit cost assumption
category_monthly['Avg_Unit_Cost'] = df.groupby(['Item_Category'])['Unit_Price'].transform('mean')

category_monthly['Dollars'] = category_monthly['Quantity'] * category_monthly['Avg_Unit_Cost']
```

```python
category_monthly['Lag_1'] = category_monthly.groupby('Item_Category')['Dollars'].shift(1)
category_monthly['Lag_2'] = category_monthly.groupby('Item_Category')['Dollars'].shift(2)
category_monthly['Rolling_3'] = category_monthly.groupby('Item_Category')['Dollars'].rolling(3).mean().reset_index(level=0, drop=True)
```

```python
category_monthly['Target'] = category_monthly.groupby('Item_Category')['Dollars'].shift(-1)
```

```python
category_monthly = category_monthly.dropna().reset_index(drop=True)
```

```python
category_monthly = pd.get_dummies(category_monthly, columns=['Item_Category'], drop_first=True)
```

```python
X = category_monthly.drop(columns=['Target', 'Month'])
y = category_monthly['Target']
```

```python
split = int(len(category_monthly) * 0.8)

X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score

print("MAE:", mean_absolute_error(y_test, y_pred))
print("R2:", r2_score(y_test, y_pred))
```
