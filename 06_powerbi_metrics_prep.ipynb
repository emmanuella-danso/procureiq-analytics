```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/cleaned/cleaned_procurement.csv")
```

```python
df.columns
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors='coerce')

df['Delivery_Date'] = pd.to_datetime(df['Delivery_Date'], errors='coerce')
```

```python
df['Month'] = df['Order_Date'].dt.to_period('M').astype(str)

df[['Order_Date', 'Month']].head()
```

```python
df['Spend'] = df['Quantity'] * df['Negotiated_Price']

df[['Quantity', 'Negotiated_Price', 'Spend']].head()
```

```python
supplier_metrics = df.groupby('Supplier').agg({
    'Spend': 'sum',
    'Delivery_Delay_Days': 'mean',
    'Defect_Rate': 'mean',
    'Compliance': lambda x: (x == 'YES').mean() * 100
}).reset_index()
```

```python
supplier_metrics.columns = [
    'Supplier',
    'Total_Spend',
    'Avg_Delivery_Delay',
    'Avg_Defect_Rate',
    'Compliance_Rate'
]

supplier_metrics.head()
```

```python
supplier_metrics.to_csv("supplier_metrics_powerbi.csv", index=False)
```

```python
category_metrics = df.groupby(['Month', 'Item_Category']).agg({
    'Spend': 'sum',
    'Total_Savings': 'sum',
    'Quantity': 'sum'
}).reset_index()
```

```python
category_metrics.columns = [
    'Month',
    'Category',
    'Total_Spend',
    'Total_Savings',
    'Total_Quantity'
]

category_metrics.head()
```

```python
category_metrics.to_csv("category_metrics_powerbi.csv", index=False)
```

```python
high_risk_suppliers = supplier_metrics[
    (supplier_metrics['Total_Spend'] > supplier_metrics['Total_Spend'].mean()) &
    (
        (supplier_metrics['Avg_Delivery_Delay'] > supplier_metrics['Avg_Delivery_Delay'].mean()) |
        (supplier_metrics['Avg_Defect_Rate'] > supplier_metrics['Avg_Defect_Rate'].mean())
    )
]
```

```python
high_risk_suppliers.to_csv("high_risk_suppliers.csv", index=False)

high_risk_suppliers.head()
```
