```python
import pandas as pd
import numpy as np
```

```python
procurement = pd.read_csv("../data/raw/Procurement KPI Analysis Dataset.csv")
```

```python
procurement_clean = procurement.copy()
```

```python
procurement_clean.dtypes
```

```python
procurement_clean['Order_Date'] = pd.to_datetime(procurement_clean['Order_Date'])

procurement_clean['Delivery_Date'] = pd.to_datetime(procurement_clean['Delivery_Date'])
```

```python
procurement_clean.info()
```

```python
procurement_clean['PO_ID'].duplicated().sum()
```

```python
procurement_clean.duplicated().sum()
```

```python
procurement_clean['Delivery_Missing'] = procurement_clean['Delivery_Date'].isnull()
```

```python
procurement_clean['Defective_Units'] = procurement_clean['Defective_Units'].fillna(0)
```

```python
procurement_clean.isnull().sum()
```

```python
text_columns = ['Supplier', 'Item_Category', 'Order_Status', 'Compliance']

for col in text_columns:
    procurement_clean[col] = procurement_clean[col].str.strip().str.upper()
```

```python
procurement_clean['Delivery_Delay_Days'] = (
    procurement_clean['Delivery_Date'] - procurement_clean['Order_Date']
).dt.days
```

```python
procurement_clean['Savings_Per_Unit'] = (
    procurement_clean['Unit_Price'] - procurement_clean['Negotiated_Price']
)

procurement_clean['Total_Savings'] = (
    procurement_clean['Savings_Per_Unit'] * procurement_clean['Quantity']
)
```

```python
procurement_clean['Defect_Rate'] = (
    procurement_clean['Defective_Units'] / procurement_clean['Quantity']
)
```

```python
procurement_clean.head()
```

```python
procurement_clean.info()
```

```python
procurement_clean.to_csv(
    "../data/cleaned/cleaned_procurement.csv",
    index=False
)
```

```python
purchases = pd.read_csv("../data/raw/purchases.csv")
```

```python
purchases.head()
```

```python
purchases.info()
```

```python
purchases.shape
```

```python
purchases.isnull().sum()
```

```python
purchases.describe()
```

```python
purchases_clean = purchases.copy()
```

```python
date_columns = ['PODate', 'ReceivingDate', 'InvoiceDate', 'PayDate']

for col in date_columns:
    purchases_clean[col] = pd.to_datetime(purchases_clean[col])
```

```python
purchases_clean.info()
```

```python
purchases_clean['Size'] = purchases_clean['Size'].fillna('UNKNOWN')
```

```python
text_columns = ['Description', 'VendorName', 'Size']

for col in text_columns:
    purchases_clean[col] = purchases_clean[col].str.strip().str.upper()
```

```python
purchases_clean.duplicated().sum()
```

```python
purchases_clean['Lead_Time_Days'] = (
    purchases_clean['ReceivingDate'] - purchases_clean['PODate']
).dt.days
```

```python
purchases_clean['Invoice_Processing_Days'] = (
    purchases_clean['InvoiceDate'] - purchases_clean['ReceivingDate']
).dt.days
```

```python
purchases_clean['Payment_Cycle_Days'] = (
    purchases_clean['PayDate'] - purchases_clean['InvoiceDate']
).dt.days
```

```python
purchases_clean['Unit_Cost'] = (
    purchases_clean['Dollars'] / purchases_clean['Quantity']
)
```

```python
purchases_clean.head()
```

```python
purchases_clean.info()
```

```python
purchases_clean.to_csv(
    "../data/cleaned/cleaned_purchases.csv",
    index=False
)
```
