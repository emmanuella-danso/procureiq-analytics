```python
import pandas as pd
import numpy as np
```

```python
procurement = pd.read_csv(
    "../data/cleaned/cleaned_procurement.csv"
)

purchases = pd.read_csv(
    "../data/cleaned/cleaned_purchases.csv"
)
```

```python
procurement.head()
```

```python
purchases.head()
```

```python
top_suppliers = (
    purchases.groupby('VendorName')['Dollars']
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

top_suppliers.head(10)
```

```python
total_spend = purchases['Dollars'].sum()

top10_spend = top_suppliers.head(10)['Dollars'].sum()

top10_percentage = (top10_spend / total_spend) * 100

print("Total Spend:", round(total_spend, 2))
print("Top 10 Supplier Spend:", round(top10_spend, 2))
print("Top 10 Spend Percentage:", round(top10_percentage, 2), "%")
```

```python
supplier_leadtime = (
    purchases.groupby('VendorName')['Lead_Time_Days']
    .mean()
    .sort_values(ascending=False)
    .reset_index()
)

supplier_leadtime.head(10)
```

```python
supplier_risk = (
    purchases.groupby('VendorName')
    .agg({
        'Dollars': 'sum',
        'Lead_Time_Days': 'mean'
    })
    .reset_index()
)

supplier_risk = supplier_risk.sort_values(
    by=['Dollars', 'Lead_Time_Days'],
    ascending=[False, False]
)

supplier_risk.head(10)
```

```python
category_savings = (
    procurement.groupby('Item_Category')['Total_Savings']
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

category_savings
```

```python
supplier_defects = (
    procurement.groupby('Supplier')['Defect_Rate']
    .mean()
    .sort_values(ascending=False)
    .reset_index()
)

supplier_defects
```

```python
category_defects = (
    procurement.groupby('Item_Category')['Defect_Rate']
    .mean()
    .sort_values(ascending=False)
    .reset_index()
)

category_defects
```

```python
supplier_payment_cycle = (
    purchases.groupby('VendorName')['Payment_Cycle_Days']
    .mean()
    .sort_values(ascending=False)
    .reset_index()
)

supplier_payment_cycle.head(10)
```

```python
supplier_unit_cost = (
    purchases.groupby('VendorName')['Unit_Cost']
    .mean()
    .sort_values(ascending=False)
    .reset_index()
)

supplier_unit_cost.head(10)
```

```python
purchases['PODate'] = pd.to_datetime(purchases['PODate'])

monthly_spend = (
    purchases.groupby(purchases['PODate'].dt.to_period('M'))['Dollars']
    .sum()
    .reset_index()
)

monthly_spend['PODate'] = monthly_spend['PODate'].astype(str)

monthly_spend
```

```python
category_spend = (
    procurement.groupby('Item_Category')[['Negotiated_Price', 'Quantity']]
    .apply(lambda x: (x['Negotiated_Price'] * x['Quantity']).sum())
    .sort_values(ascending=False)
    .reset_index(name='Total_Spend')
)

category_spend
```

```python
supplier_spend = (
    purchases.groupby('VendorName')['Dollars']
    .sum()
    .reset_index(name='Total_Spend')
)

supplier_lead = (
    purchases.groupby('VendorName')['Lead_Time_Days']
    .mean()
    .reset_index(name='Avg_Lead_Time')
)

supplier_defect = (
    procurement.groupby('Supplier')['Defect_Rate']
    .mean()
    .reset_index(name='Avg_Defect_Rate')
)

# standardize names for merge
supplier_defect['Supplier'] = supplier_defect['Supplier'].str.upper()
supplier_spend['VendorName'] = supplier_spend['VendorName'].str.upper()
supplier_lead['VendorName'] = supplier_lead['VendorName'].str.upper()

supplier_risk_final = (
    supplier_spend
    .merge(supplier_lead, on='VendorName', how='left')
    .merge(
        supplier_defect,
        left_on='VendorName',
        right_on='Supplier',
        how='left'
    )
)

supplier_risk_final = supplier_risk_final.sort_values(
    by=['Total_Spend', 'Avg_Lead_Time'],
    ascending=[False, False]
)

supplier_risk_final.head(10)
```

```python
monthly_spend['Previous_Month_Spend'] = monthly_spend['Dollars'].shift(1)

monthly_spend['Rolling_3Month_Avg'] = monthly_spend['Dollars'].rolling(3).mean()

monthly_spend['Spend_Growth_Rate'] = monthly_spend['Dollars'].pct_change()
```

```python
monthly_spend_model = monthly_spend.dropna().reset_index(drop=True)

monthly_spend_model
```

```python
monthly_spend_model.to_csv("../data/cleaned/monthly_spend_features.csv", index=False)
```
